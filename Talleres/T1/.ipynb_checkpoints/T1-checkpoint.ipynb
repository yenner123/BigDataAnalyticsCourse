```python
from bs4 import BeautifulSoup

file_text = open("Original_Docs/reut2-001.sgm","r") 
data = BeautifulSoup(file_text.read(), "html.parser")

for reuters in data.find_all('reuters'):   
    print("...")    
    print(reuters.title.get_text())
    print(reuters.body.get_text())
    print("...")    
    


print("...")    


```
